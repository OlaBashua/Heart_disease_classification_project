```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Building a Decision tree Model

```python
# spliting dataset  into training and testing sets respectively.
# the dataset was split 70:30 . 70% for training set (X_train, y_train). while 30% for testing set ( X_test ,y_test)
X = HD.drop('target', axis=1)  # Features (all columns except the target)
y = HD['target']  # Target variable (Dependent variable)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# decision tree model 
De_model = DecisionTreeClassifier()
```

```python
# Fitting the model on the training data set
De_model.fit(X_train, y_train)
```

```python
# Making  predictions on the test set
De_pred = De_model.predict(X_test)
```

```python
# Calculating the confusion matrix
De_cm = confusion_matrix(y_test, De_pred)
print("Decision Tree Confusion Matrix:")
print(De_cm)
```

```python
# Calculating the accuracy score
De_accuracy = accuracy_score(y_test, De_pred)
print("Decision Tree Accuracy:", De_accuracy)

```

```python
# Visualizing  the decision tree 
plt.figure(figsize=(15, 10))
tree.plot_tree(De_model, feature_names=X.columns, class_names=['0', '1'], filled=True)
plt.show()
```

###### Building the Random Forest Model

```python
# Building  the random forest model and 
Ra_model = RandomForestClassifier()
```

```python
#Fitting the model on the training set
Ra_model.fit(X_train, y_train)
```

```python
# Making  predictions on the test set
Ra_pred = Ra_model.predict(X_test)
```

```python
# Calculate the confusion matrix
Ra_cm = confusion_matrix(y_test, Ra_pred)
print("Random Forest Confusion Matrix:")
print(Ra_cm)
```

```python
# Calculate the accuracy score
Ra_accuracy = accuracy_score(y_test, Ra_pred)
print("Random Forest Accuracy:", Ra_accuracy)
```

```python
# Visualizing a single decision tree from the random forest
plt.figure(figsize=(15, 10))
tree.plot_tree(Ra_model.estimators_[0], feature_names=X.columns, class_names=['0', '1'], filled=True)
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
# Print the confusion matrix of all classifiers
print("Decision Tree Confusion Matrix:")
print(De_cm)
print("\nRandom Forest Confusion Matrix:")
print(Ra_cm)
```

```python
# Print the classification report of all classifiers
print("Decision Tree Classification Report:")
print(classification_report(y_test, De_pred))
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, Ra_pred))
```

```python
# Calculate the Recall Precision and F1 score of all the models
De_report = classification_report(y_test, De_pred, output_dict=True)
Ra_report = classification_report(y_test, Ra_pred, output_dict=True)
```

```python
Ra_recall = Ra_report['1']['recall']
Ra_precision = Ra_report['1']['precision']
Ra_f1 = Ra_report['1']['f1-score']
```

```python
# Visualizing the  confusion matrix using heatmaps
sns.heatmap(De_cm, annot=True, fmt='.0f', cmap='Blues')
plt.title("Decision Tree Confusion Matrix")
plt.show()

sns.heatmap(Ra_cm, annot=True, fmt='.0f', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()

# Select the best model based on the best accuracies
best_model = 'Decision Tree' if De_accuracy > Ra_accuracy else 'Random Forest'
print("Best Model:", best_model)
```
